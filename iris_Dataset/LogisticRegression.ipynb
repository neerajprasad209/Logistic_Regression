## Transform species into ordinal encoder

```python
from sklearn.preprocessing import OrdinalEncoder

encoder = OrdinalEncoder()
```

```python
encode = encoder.fit_transform(df[['species']])
```

```python
df['species'] = pd.DataFrame(encode,columns=encoder.get_feature_names_out())
```

```python
df.head()
```

```python
X = df.iloc[:,:-1]
y = df.iloc[:,-1]
```

## Spliting Independent and Dependent Variable

```python
from sklearn.model_selection import train_test_split

X_train,X_test,ytrain,ytest = train_test_split(X,y, test_size=0.20, random_state=20)
```

```python
X_train.shape, X_test.shape
```

```python
ytrain.shape, ytest.shape
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(multi_class='multinomial',solver='lbfgs')
```

```python
model.fit(X_train,ytrain)
```

```python
y_pred = model.predict(X_test)
```

```python
ytest
```

```python
## Confuaion Metrics Accuracy , Classificaton report

from sklearn.metrics import confusion_matrix, accuracy_score, classification_report

print(confusion_matrix(y_pred,ytest))
print(accuracy_score(y_pred,ytest))
print(classification_report(y_pred,ytest))
```
